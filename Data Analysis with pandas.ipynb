### map() ,apply(), applymap()

```python
titanic = pd.read_csv('Titanictrain.csv')
titanic.head()
```

```python
titanic['sex_num'] = titanic.Sex.map({'male':0,'female':1})
```

```python
titanic['name_length'] = titanic.Name.apply(len)
```

```python
titanic.loc[:,['Name','Sex','sex_num','name_length']]
```

```python
titanic.Name.str.split(',').head()
```

```python
def get_element(mylist,position):
    return mylist[position]
```

```python
titanic.Name.str.split(',').apply(get_element,position = 0)
```

```python
titanic.Name.str.split(',').apply(lambda x:x[0]).head()
```
